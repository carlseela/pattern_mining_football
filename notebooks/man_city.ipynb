```python
AREA_STEM = "England"

all_england_matches = pmf.load_matches(AREA_STEM)
```

```python

teams = pmf.load_teams()
man_city_row = next(
    t
    for t in teams
    if t.get("officialName") == "Manchester City FC"
    or t.get("name") == "Manchester City"
)
MANCHESTER_CITY_WY_ID = int(man_city_row["wyId"])


def _team_ids_in_match(match: dict) -> set[int]:
    return {int(td["teamId"]) for td in match["teamsData"].values()}


man_city_matches = [
    m for m in all_england_matches if MANCHESTER_CITY_WY_ID in _team_ids_in_match(m)
]

print([m["wyId"] for m in man_city_matches[:4]])

len(all_england_matches), len(man_city_matches), MANCHESTER_CITY_WY_ID
```

```python
# One `events_<AREA_STEM>.json` load; keeps sequences whose first ball-touching event is Man City.
man_city_ball_sequences = pmf.ball_sequences_for_team_across_matches(
    MANCHESTER_CITY_WY_ID,
    man_city_matches,
    AREA_STEM,
)

len(man_city_ball_sequences)
```
